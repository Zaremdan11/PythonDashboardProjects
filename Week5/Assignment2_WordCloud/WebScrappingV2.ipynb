```python

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
import numpy as np


res = requests.get("https://fbref.com/en/comps/9/3232/stats/2019-2020-Premier-League-Stats.html")
#next two lines take care of the breaks in the table
comm = re.compile("<!--|-->")
soup = BeautifulSoup(comm.sub("",res.text),'lxml')
all_tables = soup.findAll("tbody")
team_table = all_tables[0]
player_table = all_tables[2]

#Player Stats
pre_df_player = dict()
features_wanted_player = {"player","position","squad","age","games","games_starts","games_subs","minutes","minutes_per_game","goals","assists","pens_made","pens_att","fouls","cards_yellow","cards_red","shots_on_target","goals_per90","goals_assists_per90","goals_pens_per90","goals_assists_pens_per90","shots_on_target_per90","fouls_per90","cards_per90"}
rows_player = player_table.find_all('tr')
for row in rows_player:
    if(row.find('th',{"scope":"row"}) != None):
        for f in features_wanted_player:
            cell = row.find("td",{"data-stat": f})
            
            if cell != None:
                a = cell.text.strip().encode()
                text=a.decode("utf-8")
                if f in pre_df_player:
                    pre_df_player[f].append(text)
                else:
                    pre_df_player[f] = [text]
        df_player = pd.DataFrame.from_dict(pre_df_player)


df_player['year'] = 2019

df_player = df_player[["year","squad","player","position","age","games","games_starts","minutes","goals","assists","pens_made","pens_att","cards_yellow","cards_red","goals_per90","goals_assists_per90","goals_pens_per90","goals_assists_pens_per90"]]
#display(df_player)
df_player.to_csv('df_player2019.csv')
```

```python
import pandas as pd

df = pd.read_csv('2019PayrollPerf.csv')

total_rows = df.count
print (total_rows)
```

```python
#Deleting the records where no payroll data is present
df = df.loc[df["BaseSalary"] != 0]

total_rows = df.count
print (total_rows)
```

```python
df.head()
```

```python
#Deleting the records where no payroll data is present
df.fillna(0, inplace=True)
df.head()
```

```python
df.tail()
```

```python
indexNames = df[(df['Position'] == 'F') & (df['F_shots_on_target_pct'] == 0.0)].index

df.drop(indexNames , inplace=True)
df.shape
```

```python
indexNames = df[(df['Position'] == 'M') & (df['M_passes'] == 0)].index

df.drop(indexNames , inplace=True)
df.shape
```

```python
indexNames = df[(df['Position'] == 'D') & (df['D_tackles_won_pct'] == 0.000000)].index

df.drop(indexNames , inplace=True)
df.shape
```

```python
indexNames = df[(df['Position'] == 'GK') & (df['GK_Score'] == 0.0)].index

df.drop(indexNames , inplace=True)
df.shape
```

```python
#Normalize Defense fields
df['D_tackles_won_pct_Norm'] = df['D_tackles_won_pct'] / df.groupby('Position')['D_tackles_won_pct'].transform('sum')
#df['D_tackles_won_pct_Norm'] = 1 - df['D_tackles_won_pct_Norm']

df['D_interceptions_Norm'] = df['D_interceptions'] / df.groupby('Position')['D_interceptions'].transform('sum')
#df['D_interceptions_Norm'] = 1 - df['D_interceptions']

df['D_clearances_Norm'] = df['D_clearances'] / df.groupby('Position')['D_clearances'].transform('sum')
#df['D_clearances_Norm'] = 1 - df['D_clearances']

df['D_pressure_regain_pct_Norm'] = df['D_pressure_regain_pct'] / df.groupby('Position')['D_pressure_regain_pct'].transform('sum')
#df['D_pressure_regain_pct_Norm'] = 1 - df['D_pressure_regain_pct']

#Normalize Midfielder fields
df['M_passes_Norm'] = df['M_passes'] / df.groupby('Position')['M_passes'].transform('sum')
#df['M_passes_Norm'] = 1 - df['M_passes_Norm']

df['M_passes_pct_Norm'] = df['M_passes_pct'] / df.groupby('Position')['M_passes_pct'].transform('sum')
#df['M_passes_pct_Norm'] = 1 - df['M_passes_pct']

#Normalize Forward fields
df['F_goals_Norm'] = df['F_goals'] / df.groupby('Position')['F_goals'].transform('sum')
#df['F_goals_Norm'] = 1 - df['F_goals']

df['F_shots_on_target_pct_Norm'] = df['F_shots_on_target_pct'] / df.groupby('Position')['F_shots_on_target_pct'].transform('sum')
#df['F_shots_on_target_pct_Norm'] = 1 - df['F_shots_on_target_pct_pct']

df.head()
```

```python
df.to_csv('2019_PayrollPerfNormal.csv')
```

```python
df['D_Score'] = (df['D_clearances_Norm'] * .5) + (df['D_pressure_regain_pct_Norm'] * .3) + (df['D_tackles_won_pct_Norm'] * .2)

df['M_Score'] = (df['M_passes_Norm'] * .7) + (df['M_passes_pct_Norm'] * .3) 

df['F_Score'] = (df['F_goals_Norm'] * .7) + (df['F_shots_on_target_pct_Norm'] * .3)

```

```python
df.to_csv('2019_PayrollPerfNormalnew.csv')
```

```python
df['Position'] = df['Position'].astype(str)
```

```python
import numpy as np
#Case Statement
conditions = [
    (df['Position'] == 'M'),
    (df['Position'] == 'D'),
    (df['Position'] == 'F'),
    (df['Position'] == 'GK')
    ]

# create a list of the values we want to assign for each condition
values = [(df['M_Score'] * .6) + (df['D_Score'] * .1) + (df['F_Score'] * .3), (df['M_Score'] * .2) + (df['D_Score'] * .5) + (df['F_Score'] * .3), (df['M_Score'] * .3) + (df['F_Score'] * .7), (df['GK_Score'])]

# create a new column and use np.select to assign values to it using our lists as arguments
df['Player_Score'] = np.select(conditions, values)
```

```python
df_F = df.loc[df['Position'] == 'F']
df_F['BaseNum_Mean'] = df_F["BaseNum"].mean()
df_F['Player_Score_Mean'] = df_F["Player_Score"].mean()
df_F = df_F[["Players","Team","Position","BaseNum","Player_Score",'BaseNum_Mean','Player_Score_Mean']]
df_F['PlayerF_Salary_STD'] = (df_F['BaseNum']-df_F["BaseNum"].mean())/ (df_F["BaseNum"].std())
df_F['PlayerF_Score_STD'] = (df_F['Player_Score']-df_F["Player_Score"].mean())/ (df_F["Player_Score"].std())
df_F.info()
```
